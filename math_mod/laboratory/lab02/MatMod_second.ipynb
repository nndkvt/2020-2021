```python
import matplotlib.pyplot as plt
import math
from scipy.integrate import odeint
import numpy as np
```

```python
s = 16.2
fi = 3*math.pi / 4
```

```python
def model(r, tetha):
    dr = r / math.sqrt(3)
    return dr
```

```python
r0 = s
tetha0 = -math.pi
tetha = np.arange(tetha0, 2*math.pi, .01)
r = odeint(model, r0, tetha)
```

```python
def f2(t):
    xt = math.tan(fi)*t
    return xt
```

```python
# точка пересечения на графике лодки

t2 = math.sqrt((r[864]**2)/2)
t2
```

```python
t = np.arange(0, int(t2+t2*.25), 1)
```

```python
# траектория катера береговой охраны

plt.polar(tetha, r)
```

```python
# траектория лодки браконьеров
# красная точка - пересечение с траекторией катера

plt.plot(t, f2(t))
plt.plot([t2, t2+.01], [-t2, -t2-.01], 'r', linewidth=5)
```
