```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from scipy.integrate import odeint
```

```python
a = .18
b = .032
c = .19
d = .026
```

```python
def syst(yf, t):
    x = yf[0]
    y = yf[1]
    dxdt = a*x - b*x*y
    dydt = -c*y + d*x*y
    return (dxdt, dydt)
```

```python
x0 = 8
y0 = 3
```

```python
x_st = c/d
y_st = a/b
```

```python
t0 = 0
tmax = 400
dt = .1

t = np.arange(t0, tmax, dt)
t = np.append(t, tmax)
t
```

```python
v0 = (x0, y0)
```

```python
yf = odeint(syst, v0, t)
yf
```

```python
x = []
y = []

for i in range(len(yf)):
    x = np.append(x, yf[i][0])
    y = np.append(y, yf[i][1])
```

```python
plt.figure(figsize=(12.8, 9.6))
plt.grid(True)
plt.title('Модель "хищник-жертва"', fontsize=25)
plt.xlabel("Жертвы", fontsize=20)
plt.ylabel("Хищники", fontsize=20)
plt.plot(x, y)
plt.plot([x_st], [y_st], 'ro', label='Стац. состояние')
plt.legend(loc='upper right', fontsize=15)
```

```python
plt.figure(figsize=(12.8, 9.6))
plt.grid(True)
plt.title('Модель "хищник-жертва"', fontsize=25)
plt.xlabel("t", fontsize=20)
plt.plot(t, x, 'g', label='Жертвы')
plt.plot(t, y, 'r', label='Хищники')
plt.legend(loc='upper right', fontsize=15)
```
