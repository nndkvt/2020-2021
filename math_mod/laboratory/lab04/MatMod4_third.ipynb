```python
import numpy as np
import pandas as pd
import scipy as sp
import math
import matplotlib.pyplot as plt

from scipy.integrate import odeint
```

```python
x0 = 2
y0 = -1
tmax = 60
t0 = 0
dt = 0.05
```

```python
t = np.arange(t0, tmax, dt)
t = np.append(t, tmax)
t
```

```python
w = math.sqrt(3.0)
g = 12.0 / 2
```

```python
def f(t):
    return (0.2 * math.sin(5*t))
```

```python
def func(y, t):
    dxdt = y[1]
    dydt = -w*w*y[0] - g*2*y[1] - f(t)
    return (dxdt, dydt)
```

```python
v0 = (x0, y0)
```

```python
yf = odeint(func, v0, t)
yf
```

```python
x = []
y = []
for i in range(len(yf)):
    x = np.append(x, yf[i][0])
    y = np.append(y, yf[i][1])
```

```python
plt.figure(figsize=(12.8, 9.6))
plt.grid(True)
plt.title("Колебания гармонического осциллятора c затуханием\n и под действием внешней силы", fontsize=25)
plt.xlabel("x", fontsize=20)
plt.ylabel("y", fontsize=20)
plt.plot(x, y)
```
